```python
# The numpy array operations are faster
L = range(1000)

# We time a function to create another list which has each number squared using list comprehension
print "List comprehension"
%timeit [i**2 for i in L]

# We also perform the same operation using array
import numpy as np
print "Array operation"
a = np.array(L)
%timeit a**2
```

```python
# Array operation are about 30 times faster
```

### 4) Broadcasting

```python
# The idea of broadcasting allows us to add or perform elementwise type operations on arrays 
# even though they are of different sizes.
# Example

a = np.tile(np.arange(0, 40, 10), (3,1)).T
b = np.arange(0,3)
# Although the two arrays are not of the same size but can still be added
print a+b
```

```python
# Working example
# Calculating the distances between various cities along Route66
mileposts = np.array([0, 198, 303, 736, 871, 1175, 1475, 1544, 1913, 2448])
print "mileposts: ", mileposts
mileposts_across = mileposts[:, np.newaxis]
print "mileposts_across: ", mileposts_across
```

```python
# We can obtain a distance table directly by subtracting the above arrays
distance = np.abs(mileposts-mileposts_across)
print distance
```

### 5) Array shape manipulation

```python
# Flattening
a = np.array([[1,2,3], [4,5,6]])
print 'a.ravel()', a.ravel()
print 'a.T', a.T
print 'a.T.ravel()', a.T.ravel()
```

```python
# Reshaping
b = a.ravel()
print 'b: ', b
reshaped_b = b.reshape(3,2)
print 'reshaped_b: ', reshaped_b
```

## 6) Sorting

```python
# Sorting a 2D array
a = np.array([[4, 3, 5], [1, 2, 1]])
b = np.sort(a, axis=1) # axis =1 means the sorting will be done along each row through the columns
print 'a: ', a
print 'b: ', b
# sorting a on its own
a.sort(axis=1)
print 'a: ', a
```

```python
# Argsort returns an array of indexes of the array in the increasing order of their elements
a = np.array([2,5,1,6,4])
print 'a: ', a
arg_s = np.argsort(a)
print 'arg_s: ', arg_s
print 'a[arg_s]: ', a[arg_s]
```

## 7) Casting

```python
a = np.array([1.3, 2.22])
print 'a: ', a
print 'a.astype(int): ', a.astype(int)
```
