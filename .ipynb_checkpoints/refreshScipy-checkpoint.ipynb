## 3) Fast Fourier Transform

```python
# Here we study a basics of fast fourier transform
t_step = 0.02
period = 5.

t_vec = np.arange(0, 20, t_step)
sig = (np.sin(2 * np.pi / period * t_vec) + \
       0.5 * np.random.randn(t_vec.size))
plt.plot(t_vec, sig, 'b-')
plt.title('Signal vs t_vec')
plt.show()
```

```python
# Performing FFT
from scipy import fftpack

# Creating a set of sampling frequencies
sample_freq = fftpack.fftfreq(sig.size, d=t_step)
# Performing the fast fourier transform
sig_fft = fftpack.fft(sig)
#print sig_fft
```

```python
# Extracting the sample frequencies > 0
pidxs = np.where(sample_freq > 0)
freqs = sample_freq[pidxs]

# Evaluating power for the fourier transform for freq > 0
power = np.abs(sig_fft)[pidxs]
freq = freqs[power.argmax()]
np.allclose(freq, 1./period)  # check that correct freq is found
sig_fft[np.abs(sample_freq) > freq] = 0
main_sig = fftpack.ifft(sig_fft)
```

```python
# Plotting the filtred data
plt.figure()    
plt.plot(t_vec, sig)
plt.plot(t_vec, main_sig, linewidth=3)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

## Denoising the moonlanding image

```python
# Looking at the original image
from pylab import imread
moon_Im = imread('data/moonlanding.png')
plt.imshow(moon_Im, cmap=plt.cm.gray)
print moon_Im.min()
plt.show()
```

```python
# Checking the size of moon_Im
print 'Dimension of moon_Im: \n', moon_Im.shape
```

```python
# Creating a set of sampling frequencies
moon_fft = fftpack.fft2(moon_Im)
print moon_fft.shape
```

```python
# Remving high frequency signlas
f_tr = 50
moon_fft_clean = moon_fft.copy()

moon_fft_clean[:, f_tr:-1*f_tr] = 0
moon_fft_clean[f_tr:-1*f_tr, :] = 0

# Inverse fast fourier transform
moon_new = np.fft.ifft2(moon_fft_clean).real
plt.imshow(moon_new, plt.cm.gray)
plt.show()
```

## 4) Optimizing

#### a) Finding minima

```python
# This section is to learn about basic data fitting
from scipy import optimize

# A basic function to define the variables
def f(x):
    return x**2 + 10*np.sin(x)

# plotting the function over a range of x values
x = np.arange(-10, 10, 0.1)
plt.plot(x, f(x)) 
plt.show() 
```

```python
# We wish to find the minima for this function
# We use the basic fmin function with gradient seach and initial guess 0
optimize.fmin_bfgs(f, 0)
```

```python
# However if we initiate the above with a different value, it will lead to the local minima
optimize.fmin_bfgs(f, 3)
```

```python
# There is a better basinhopping algorithm that finds the global minima
optimize.basinhopping(f,0)
```

```python
optimize.basinhopping(f,5)
```

#### Basin hopping finds the global minima even though we started on extreme right

#### b) FInding Coefficients

```python
x_data = np.linspace(-10, 10, 21)
# We generate a y_data using the function defined previously and adding some random noise
y_data = f(x_data) + np.random.randn(x_data.size)
```

```python
# So we know the functional form that we want to fit but need to evaluate the coefficients
def f2(x, a, b):
    return a*(x**2) + b*np.sin(x)
```

```python
# Initial guess for the parameters
initial_guess = [3,3]
params, param_cov = optimize.curve_fit(f2, x_data, y_data, initial_guess)
print params
```

### Exercise 2d- minimization

```python
# Given the function
def f_2d(x):
    return ((4 - 2.1*(x[0]**2) + (x[0]**4)/3.)*(x[0]**2) + x[0]*x[1] + 4*(x[1]**2 - 4)*(x[1]**2))

x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)
x_grid, y_grid = np.meshgrid(x, y)
z_f2d = f_2d([x_grid, y_grid])

plt.imshow(z_f2d, cmap=plt.cm.jet, extent=[-2, 2, -2, 2])
plt.colorbar()
plt.show()
```

### 5) Statistics

```python
a = np.random.normal(size=1000)
#print a
bins = np.arange(-4, 5, 1)

# Distributing the randomly generated numbers in a into bins
histogram = np.histogram(a, bins=bins, normed=True)[0]
bins = 0.5*(bins[1:] + bins[:-1])

# Using stats package in scipy to obtain the normal funtion
from scipy import stats
b = stats.norm.pdf(bins)

# Plotting the two funtions
plt.plot(bins, histogram)
plt.plot(bins, b)
```

#### A simple maximum likelihood estimate based parameter set for an assumed function can be obtained using the fit function as follows

```python
# We have a distribution of random variables in 'a'. To obtain the mean and the std dev we use the funtion fit
mu, std = stats.norm.fit(a)
print 'mu: ',mu
print 'std: ',std
```

```python
# Percentile values can be obtained using scoreatpercentile funtion
print stats.scoreatpercentile(a,96)
```

###  6) Calculus

```python
# Integrate
from scipy.integrate import quad
res, error = quad(np.sin, 0, (np.pi/2))
np.allclose(res, 1)
```

```python
# Solving simple differential equations like dy/dt = -2y
# We first need to define a function that calculates the derivative
def calc_derivative(ypos, time):
    return -2*ypos

# Module to obtain trajectory
from scipy.integrate import odeint
time_vec = np.linspace(0, 10, 100)
y_vec, info = odeint(calc_derivative, 1, time_vec, full_output=True)

# Plotting the y_vec
plt.plot(time_vec, y_vec)
plt.xlabel('time')
plt.ylabel('y')
plt.show()
```
